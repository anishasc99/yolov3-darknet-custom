```
import subprocess
import os

class yolo:
  def __init__(self):
    subprocess.Popen(["cd", "/content/yolov3-darknet-custom"] ,shell=True, 
                                  stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    return 
  def train(self, data = "data/yolo.data",cfg = "cfg/yolov3_custom_train.cfg",weights = "darknet53.conv.74"):
    #python generate.py
    print("Training")
    cmd_list = ['python', 'generate.py']
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print(result.stderr.decode())
    print(result.stdout.decode())

    #./darknet detector train data/yolo.data cfg/yolov3_custom_train.cfg darknet53.conv.74 
    cmd_list = ['./darknet', 'detector', 'train', data, cfg, weights]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print(result.stderr.decode())
    print(result.stdout.decode())

    #python tensorboard_generate.py
    cmd_list = ['python', 'tensorboard.py']
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print(result.stderr.decode())
    print(result.stdout.decode())
    return 
  def test(self,model = 'weights/vgg16_reducedfc.pth'):
    #python test.py --trained_model "weights/COCO (1).pth"
    result = subprocess.run(['python','test.py','--trained_model',model], stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Testing")
    print( result.stderr.decode())
    print (result.stdout.decode())
    result = subprocess.run(['python','coco_eval.py'], stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("COCO Evaluation")
    print( result.stderr.decode())
    print (result.stdout.decode())
    return
  def test_one(self, data = "data/yolo.data", cfg = "cfg/yolov3_custom_train.cfg", weights = "/content/yolov3-darknet-custom/yolov3_custom_train_1000.weights", image = "data/11003-Ravi-Singh-Baroda-Gujarat-1.jpg", thresh = "0.25"):
    
    # ./darknet detector test data/yolo.data cfg/yolov3_custom_train.cfg '/content/yolov3-darknet-custom/backup/yolov3_custom_train_1000.weights' 'data/11003-Ravi-Singh-Baroda-Gujarat-1.jpg' --thresh 0.25
    cmd_list = ['./darknet', 'detector', 'test', data, cfg, weights, image, thresh]
    result = subprocess.run(cmd_list, stdout=subprocess.PIPE, stderr = subprocess.PIPE)
    print("Testing sample image")
    print(result.stderr.decode())
    print (result.stdout.decode())
    return
```
